```python
X = df_reduced.drop(columns=['SECONDVISITFLAG'])
y = df_reduced['SECONDVISITFLAG'].map({'Y': 1, 'N': 0})
```

### missing values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

### Standardize the data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

### handling imbalance using SMOTE

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)
```

### Splitting the data

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

### XGBoost model

```python
xgb_model = XGBClassifier(random_state=42, eval_metric='logloss')
```

### the parameter grid for RandomizedSearchCV

```python
param_dist_xgb = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [3, 4, 5, 6, 7],
    'learning_rate': [0.01, 0.02, 0.04],
    'subsample': [1.0, 0.4],
    'colsample_bytree': [0.7, 1.0]
}
```

### Randomized Search

```python
random_search_xgb = RandomizedSearchCV(estimator=xgb_model, param_distributions=param_dist_xgb, n_iter=50, cv=3, random_state=42, n_jobs=-1, scoring='f1')
random_search_xgb.fit(X_train, y_train)
```

### Best parameters and model

```python
print(f'Best parameters: {random_search_xgb.best_params_}')
best_xgb_model = random_search_xgb.best_estimator_
```

### predictions

```python
y_pred_xgb = best_xgb_model.predict(X_test)
```

### Evaluating the model

```python
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
precision_xgb = precision_score(y_test, y_pred_xgb)
recall_xgb = recall_score(y_test, y_pred_xgb)
f1_xgb = f1_score(y_test, y_pred_xgb)
roc_auc_xgb = roc_auc_score(y_test, y_pred_xgb)
```
